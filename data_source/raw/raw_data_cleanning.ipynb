## PAÍS

```python
# Coleta de dados globais do Brasil
df_reg_pais = df[df['regiao'] == 'Brasil']
df_reg_pais
```

```python
# Exclusão das colunas de dados relacionadas a estados e municipios
df_reg_pais = df_reg_pais.drop(['estado', 'municipio', 'codmun', 'codRegiaoSaude', 'nomeRegiaoSaude', 'interior/metropolitana'], axis=1)
```

## ESTADO

```python
# Coleta de dados globais dos estados
df_reg_estados = df.iloc[158:4424]
df_reg_estados
```

```python
df_reg_estados.info()
```

```python
# Exclusão das colunas de dados relacionadas a municipios e colunas com dados completamente nulos
df_reg_estados = df_reg_estados.drop(['municipio', 'codmun', 'codRegiaoSaude', 'nomeRegiaoSaude', 'Recuperadosnovos', 'emAcompanhamentoNovos', 'interior/metropolitana'], axis=1)
```

```python
# Correção de tipos floats para inteiros
df_reg_estados['populacaoTCU2019'] = df_reg_estados['populacaoTCU2019'].astype(int)
```

```python
# Renomeação de coluna
df_reg_estados = df_reg_estados.rename(columns={'semanaEpi': 'semanaEpidemiologica'})
```

```python
df_reg_estados
```

```python
# Verificação de dados limpos
df_reg_estados.info()
```

```python
# Lista de estados
df_reg_estados['estado'].unique()
```

```python
# Criação de um dicionário de estados SEM CONSIDERAR O PERÍODO DE ANÁLISE
estados = {}
for estado in df_reg_estados['estado'].unique():
    estados[estado] = df_reg_estados[df_reg_estados['estado'] == estado]
```

```python
estados['RO']
```

```python
# O Período completo da parte 1 dos dados tem 158 dias do dia 25 Fev a 31 Jul
# O Período de análise do projeto tem 87 dias e vai do dia 14 Mar a 08 Jun
# O código a seguir irá extrair o intervalo de análise de cada estado
# E irá concatenar estes intervalos de dados no DF df_reg_estados_periodo
periodo_completo = 158
inicio_analise = 18  # dia 14 Mar
fim_analise = 105  # dia 08 Jul
n_estados = 27

intervalos_concatenados = []
for inter in range(n_estados):
    intervalos_concatenados.append(df_reg_estados.iloc[inicio_analise:fim_analise])
    inicio_analise += periodo_completo
    fim_analise += periodo_completo

df_reg_estados_periodo = pd.concat(intervalos_concatenados)
df_reg_estados_periodo
```

```python
# Verificação: O período possui 87 dias em 27 estados, totalizando 87x27 registros no DataFrame.
if len(df_reg_estados_periodo) == 87*27:
    print('OK')
else:
    print('FAIL')
```

```python
# Cria um dicionário em que as chaves são as siglas dos estados e os valores são os intervalos do DF correspondentes aos próprios estados.
estados_periodo_dict = {}
for estado in df_reg_estados_periodo['estado'].unique():
    estados_periodo_dict[estado] = df_reg_estados_periodo[df_reg_estados_periodo['estado'] == estado]
```

```python
estados_periodo_dict['RJ']
```

```python
# Salvando todo o período da parte 1 dos dados em csv
df_reg_estados.to_csv('HIST_PAINEL_COVIDBR_2020_Parte1_23jun2023_REGIAO_ESTADOS.csv', sep=";", index=False)

# Salvando dados do período em csv
df_reg_estados_periodo.to_csv('HIST_COVIDBR_2020_PER_14MARÇO_08JUNHO_REGIAO_ESTADOS.csv', sep=";", index=False)
```
